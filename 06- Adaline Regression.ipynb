```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# from numpy.linalg import inv
```

### generate data for weight and height of humans

```python
# generate data function

def generate_data(N):
    
    Height = np.random.randint(80, 190, size=N)
    m = np.random.uniform(0.3, 0.5, N)
    b = np.random.uniform(5, 10, N)

    Weight = Height * m + b 

    return Height, Weight
```

```python
N = 100
Height, Weight =  generate_data(N)

print(Height.shape)
print(Weight.shape)
```

```python
#reshape data

X = Height.reshape(-1, 1)
Y = Weight.reshape(-1, 1)

print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X, Y, c='blue')

plt.xlabel('Height')
plt.ylabel('Weight')
```

### plot random slope

```python
M = np.random.uniform(0.5, 0.9, 3)
# print(M)

plt.scatter(X, Y,c='blue')
plt.xlabel('Height')
plt.ylabel('Weight')

for m in M:
    y_pred = np.matmul(X, [m])    
    plt.plot(X, y_pred, c='red', lw=2)
```

### Adaline - plot slope 

```python
# m = (XT. X)^-1 * XTY

m = np.matmul(np.linalg.inv(np.matmul(X.T, X)),  np.matmul(X.T, Y))
print(m)
y_pred = np.matmul(X, m)
# print(y_pred)
plt.scatter(X, Y, c='blue')
plt.plot(X, y_pred, c='magenta', lw=2)
plt.xlabel('Height')
plt.ylabel('Weight')
```

### linregress of scipy package - plot slope 

```python
from scipy.stats import linregress

result  = linregress(X[:, 0], Y[:, 0])
print(result.slope)
y_pred = result.intercept + np.matmul(X, np.array([[result.slope]]))
# print(y_pred)
plt.scatter(X, Y, c = 'blue')
plt.plot(X, y_pred, c='magenta', lw = 2)
plt.xlabel('Height')
plt.ylabel('Weight')
```
